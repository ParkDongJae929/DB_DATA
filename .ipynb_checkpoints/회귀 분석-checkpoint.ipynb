```python
#============================================================================
# [Cell 1] ë¼ì´ë¸ŒëŸ¬ë¦¬ ì„í¬íŠ¸
#============================================================================
"""
Gorton & Metrick (2012) í•œêµ­ ì ìš©

[ë³€ìˆ˜ ëŒ€ì‘]
- ì¢…ì†ë³€ìˆ˜ (ì „ì²´): KOFR - ê¸°ì¤€ê¸ˆë¦¬ â† Repo - OIS
- ì¢…ì†ë³€ìˆ˜ (íŒ¨ë„): ë‹´ë³´ë³„ RPê¸ˆë¦¬ - ê¸°ì¤€ê¸ˆë¦¬ â† Repo - OIS (ë‹´ë³´ë³„)
- BankStress: ì½œê¸ˆë¦¬ - ê¸°ì¤€ê¸ˆë¦¬ â† LIB-OIS
- CreditSpread: íšŒì‚¬ì±„(BBB-) - êµ­ê³ ì±„3Y â† ABX BBB
- VKOSPI â† VIX (í†µì œë³€ìˆ˜)

â€» NonBankStress ì œì™¸: CreditSpreadì™€ ìƒê´€ 0.942, Granger ì¸ê³¼ì„± ì—†ìŒ
"""

import pandas as pd
import numpy as np
import sqlite3
import statsmodels.api as sm
from statsmodels.regression.linear_model import OLS
from linearmodels.panel import PanelOLS
from scipy import stats
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import seaborn as sns
from pathlib import Path
import warnings

warnings.filterwarnings('ignore')

plt.rcParams['font.family'] = 'Malgun Gothic'  # Windows
plt.rcParams['axes.unicode_minus'] = False
plt.rcParams['figure.figsize'] = (12, 6)

# ========================== ì‚¬ìš©ì ì„¤ì • ==========================
# ê²½ë¡œ ì„¤ì •
rate_path = 'C:\\Users\\jay15\\Desktop\\DB_DATA\\DataBase\\ì‹œì¥ê¸ˆë¦¬(ì¼ë³„)_250109.csv'
stock_path = 'C:\\Users\\jay15\\Desktop\\DB_DATA\\DataBase\\êµ­ë‚´ì£¼ê°€ì§€ìˆ˜(ì¼ë³„)_250109.csv'
vkospi_path = 'C:\\Users\\jay15\\Desktop\\DB_DATA\\DataBase\\VKOSPI(ì¼ë³„)_251231.csv'
repo_db_path = 'C:\\Users\\jay15\\Desktop\\DB_DATA\\DataBase\\D_Repo_2015-2025.db'
repo_rate_db_path = 'C:\\Users\\jay15\\Desktop\\DB_DATA\\DataBase\\D_Repo_2015-2025.db'
```

```python
# ë¶„ì„ ê¸°ê°„
START_DATE = "20150101"
END_DATE = "20251231"

# ì‹œì°¨ ìˆ˜
N_LAGS = 3

# ì¶œë ¥ í´ë”
OUTPUT_DIR = './output'
Path(OUTPUT_DIR).mkdir(parents=True, exist_ok=True)
# ================================================================

print("=" * 70)
print("ğŸ“‚ ë°ì´í„° ë¡œë“œ ë° ì „ì²˜ë¦¬")
print("=" * 70)

# -----------------------------
# 1) ì‹œì¥ê¸ˆë¦¬ ë°ì´í„°
# -----------------------------
df_rate = pd.read_csv(rate_path, encoding='utf-8')
df_rate['date'] = pd.to_datetime(df_rate['DATE'])

# ì»¬ëŸ¼ëª… ì •ë¦¬ (ê³µë°± ì œê±°)
df_rate.columns = df_rate.columns.str.strip()

# ì»¬ëŸ¼ëª… í‘œì¤€í™”
rename_dict = {
    'ê¸°ì¤€ê¸ˆë¦¬': 'BASE_RATE',
    'CD(91ì¼)': 'CD91',
    'CP(91ì¼, A1)': 'CP91',
    'KOFR(ê³µì‹œRFR)': 'KOFR',
    'ì½œê¸ˆë¦¬(1ì¼, ì „ì²´ê±°ë˜)': 'CALL',
    'êµ­ê³ ì±„(3ë…„)': 'KTB3Y',
    'êµ­ê³ ì±„(10ë…„)': 'KTB10Y',
    'êµ­ê³ ì±„(2ë…„)': 'KTB2Y',
    'í†µì•ˆì¦ê¶Œ(91ì¼)': 'MSB91',
    'íšŒì‚¬ì±„(3ë…„, AA-)': 'CORP_AA',
    'íšŒì‚¬ì±„(3ë…„, BBB-)': 'CORP_BBB'
}
df_rate = df_rate.rename(columns=rename_dict)

print(f"  ì‹œì¥ê¸ˆë¦¬: {len(df_rate):,}ì¼")

# -----------------------------
# 2) ì£¼ê°€ì§€ìˆ˜ ë°ì´í„°
# -----------------------------
df_stock = pd.read_csv(stock_path, encoding='utf-8')
df_stock['date'] = pd.to_datetime(df_stock['DATE'])
df_stock['KOSPI'] = pd.to_numeric(df_stock['KOSPI'].astype(str).str.replace(',', ''), errors='coerce')
print(f"  ì£¼ê°€ì§€ìˆ˜: {len(df_stock):,}ì¼")

# -----------------------------
# 3) VKOSPI ë°ì´í„°
# -----------------------------
df_vkospi = pd.read_csv(vkospi_path, encoding='cp949')
df_vkospi['date'] = pd.to_datetime(df_vkospi['DATE'])
df_vkospi = df_vkospi.rename(columns={'ì¢…ê°€': 'VKOSPI'})
print(f"  VKOSPI: {len(df_vkospi):,}ì¼")

# -----------------------------
# 4) ë‹´ë³´ë³„ RP ê¸ˆë¦¬ ë°ì´í„° (ìƒˆ DB)
# -----------------------------
conn = sqlite3.connect(repo_db_path)
df_repo_daily = pd.read_sql("SELECT * FROM daily_repo_rates", conn)
conn.close()

# ë‚ ì§œ ì»¬ëŸ¼ ì²˜ë¦¬
df_repo_daily['date'] = pd.to_datetime(df_repo_daily['basDt'])
df_repo_daily = df_repo_daily.drop(columns=['basDt'])

print(f"  ë‹´ë³´ë³„ RP ê¸ˆë¦¬: {len(df_repo_daily)}ì¼, {len(df_repo_daily.columns)-1}ê°œ ë‹´ë³´ìœ í˜•")

# ë‹´ë³´ìœ í˜• í™•ì¸
collateral_cols = [col for col in df_repo_daily.columns if col != 'date']
print(f"  ë‹´ë³´ìœ í˜•: {collateral_cols}")

conn_repo_rate = sqlite3.connect(repo_rate_db_path)
df_repo_total = pd.read_sql("SELECT * FROM daily_repo_rates", conn_repo_rate)
conn_repo_rate.close()

# ë‚ ì§œ ì»¬ëŸ¼ ì²˜ë¦¬
if 'basDt' in df_repo_total.columns:
    df_repo_total['date'] = pd.to_datetime(df_repo_total['basDt'])
    df_repo_total = df_repo_total.drop(columns=['basDt'])
elif 'index' in df_repo_total.columns:
    df_repo_total['date'] = pd.to_datetime(df_repo_total['index'])
    df_repo_total = df_repo_total.drop(columns=['index'])

# 'ì „ì²´' ì»¬ëŸ¼ë§Œ ì¶”ì¶œ
df_repo_total = df_repo_total[['date', 'ì „ì²´']].rename(columns={'ì „ì²´': 'REPO_TOTAL'})
print(f"  ì „ì²´ RP ê¸ˆë¦¬: {len(df_repo_total)}ì¼")

# -----------------------------
# 5) ì‹œì¥ ë°ì´í„° ë³‘í•©
# -----------------------------
df_market = df_rate.merge(df_stock[['date', 'KOSPI']], on='date', how='outer')
df_market = df_market.merge(df_vkospi[['date', 'VKOSPI']], on='date', how='outer')
df_market = df_market.merge(df_repo_total[['date', 'REPO_TOTAL']], on='date', how='outer')  # ì¶”ê°€

df_market = df_market.sort_values('date').reset_index(drop=True)

# ê¸°ê°„ í•„í„°
start_dt = pd.to_datetime(START_DATE, format='%Y%m%d')
end_dt = pd.to_datetime(END_DATE, format='%Y%m%d')
df_market = df_market[(df_market['date'] >= start_dt) & (df_market['date'] <= end_dt)]

print(f"  ì‹œì¥ë°ì´í„° ë³‘í•©: {len(df_market):,}ì¼")

# -----------------------------
# 6) ë³€ìˆ˜ ìƒì„± (bp ë‹¨ìœ„)
# -----------------------------

# (1) ì¢…ì†ë³€ìˆ˜ (ì „ì²´ ì‹œì¥): RepoSpread = KOFR - ê¸°ì¤€ê¸ˆë¦¬
df_market['RepoSpread'] = (df_market['REPO_TOTAL'] - df_market['BASE_RATE']) * 100

# (2) BankStress = ì½œê¸ˆë¦¬ - ê¸°ì¤€ê¸ˆë¦¬ : ì€í–‰ ìœ ë™ì„± í”„ë¦¬ë¯¸ì—„
df_market['BankStress'] = (df_market['CALL'] - df_market['BASE_RATE']) * 100

# (3) CreditSpread = íšŒì‚¬ì±„(BBB-) - êµ­ê³ ì±„(3ë…„) : ì‹ ìš©ìœ„í—˜
df_market['CreditSpread'] = (df_market['CP91'] - df_market['MSB91']) * 100

# (4) YieldSlope = êµ­ê³ ì±„(10ë…„) - êµ­ê³ ì±„(2ë…„)
df_market['YieldSlope'] = (df_market['KTB10Y'] - df_market['KTB3Y']) * 100

# (5) r10 = êµ­ê³ ì±„(10ë…„) ìˆ˜ì¤€
df_market['r10'] = df_market['KTB10Y']

# (6) KOSPI ìˆ˜ìµë¥  (%)
df_market['KOSPI_ret'] = df_market['KOSPI'].pct_change() * 100

# -----------------------------
# 7) ì „ì²´ ì‹œì¥ ë¶„ì„ ë°ì´í„°
# -----------------------------
analysis_cols = ['date', 'RepoSpread', 'BankStress', 'CreditSpread', 
                 'YieldSlope', 'r10', 'VKOSPI', 'KOSPI_ret', 'BASE_RATE']
df_analysis = df_market[analysis_cols].dropna()
df_analysis = df_analysis.set_index('date').sort_index()

print(f"\n  ì „ì²´ ì‹œì¥ ë¶„ì„ ë°ì´í„°: {len(df_analysis)}ì¼")
print(f"  ê¸°ê°„: {df_analysis.index.min().strftime('%Y-%m-%d')} ~ {df_analysis.index.max().strftime('%Y-%m-%d')}")

# -----------------------------
# 8) íŒ¨ë„ ë°ì´í„° êµ¬ì„± (ë‹´ë³´ìœ í˜•ë³„)
# -----------------------------
print(f"\n{'='*70}")
print("ğŸ“Š íŒ¨ë„ ë°ì´í„° êµ¬ì„±")
print("="*70)

# ë‹´ë³´ë³„ RP ê¸ˆë¦¬ì™€ ì‹œì¥ ë°ì´í„° ë³‘í•©
df_repo_daily_filtered = df_repo_daily[
    (df_repo_daily['date'] >= start_dt) & (df_repo_daily['date'] <= end_dt)
]

# Long formatìœ¼ë¡œ ë³€í™˜
df_panel_list = []
for col in collateral_cols:
    temp = df_repo_daily_filtered[['date', col]].copy()
    temp = temp.rename(columns={col: 'RP_rate'})
    temp['collateral'] = col
    temp = temp.dropna(subset=['RP_rate'])
    df_panel_list.append(temp)

df_panel_long = pd.concat(df_panel_list, ignore_index=True)

# ì‹œì¥ ë³€ìˆ˜ ë³‘í•©
df_panel = df_panel_long.merge(
    df_market[['date', 'BASE_RATE', 'BankStress', 'CreditSpread', 
               'YieldSlope', 'r10', 'VKOSPI', 'KOSPI_ret']],
    on='date', how='inner'
)

# ë‹´ë³´ë³„ ìŠ¤í”„ë ˆë“œ ê³„ì‚° (bp)
df_panel['RepoSpread'] = (df_panel['RP_rate'] - df_panel['BASE_RATE']) * 100

# ê²°ì¸¡ì¹˜ ì œê±°
df_panel = df_panel.dropna()

# íŒ¨ë„ ì¸ë±ìŠ¤ ì„¤ì •
df_panel = df_panel.set_index(['collateral', 'date'])
df_panel = df_panel.sort_index()

print(f"  íŒ¨ë„ ë°ì´í„°: {len(df_panel):,} ê´€ì¸¡ì¹˜")
print(f"  ë‹´ë³´ìœ í˜•: {df_panel.index.get_level_values('collateral').nunique()}ê°œ")
print(f"  ê¸°ê°„: {df_panel.index.get_level_values('date').min().strftime('%Y-%m-%d')} ~ {df_panel.index.get_level_values('date').max().strftime('%Y-%m-%d')}")

# ë‹´ë³´ìœ í˜•ë³„ ê´€ì¸¡ì¹˜ ìˆ˜
print(f"\n  [ë‹´ë³´ìœ í˜•ë³„ ê´€ì¸¡ì¹˜]")
obs_by_coll = df_panel.groupby(level='collateral').size()
print(obs_by_coll)

# -----------------------------
# 9) ê¸°ì´ˆ í†µê³„
# -----------------------------
print(f"\n{'='*70}")
print("ğŸ“Š ê¸°ì´ˆ í†µê³„")
print("="*70)

print("\n[1] ì „ì²´ ì‹œì¥ (KOFR ê¸°ë°˜)]")
stats_cols = ['RepoSpread', 'BankStress', 'CreditSpread', 'VKOSPI']
stats_df = df_analysis[stats_cols].describe().T[['mean', 'std', 'min', 'max']]
stats_df.columns = ['í‰ê· ', 'í‘œì¤€í¸ì°¨', 'ìµœì†Œ', 'ìµœëŒ€']
print(stats_df.round(2))

print("\n[2] íŒ¨ë„ ë°ì´í„° (ë‹´ë³´ë³„)]")
panel_stats = df_panel.groupby(level='collateral')['RepoSpread'].agg(['mean', 'std', 'min', 'max'])
panel_stats.columns = ['í‰ê· ', 'í‘œì¤€í¸ì°¨', 'ìµœì†Œ', 'ìµœëŒ€']
print(panel_stats.round(2))

print("\n[ë³€ìˆ˜ë³„ ê²°ì¸¡ì¹˜ í˜„í™©]")
print("-" * 50)
for col in ['REPO_TOTAL', 'CALL', 'CP91', 'MSB91', 'VKOSPI', 'BASE_RATE']:
    if col in df_market.columns:
        valid = df_market[col].notna().sum()
        first_valid = df_market.loc[df_market[col].notna(), 'date'].min()
        print(f"  {col:15s}: {valid:,}ì¼, ì‹œì‘ì¼: {first_valid.strftime('%Y-%m-%d') if pd.notna(first_valid) else 'N/A'}")

# analysis_colsë³„ ê²°ì¸¡ì¹˜ í™•ì¸
print("\n[analysis_cols ë³€ìˆ˜ë³„ ê²°ì¸¡ì¹˜ í˜„í™©]")
print("-" * 60)
analysis_cols = ['RepoSpread', 'BankStress', 'CreditSpread', 
                 'YieldSlope', 'r10', 'VKOSPI', 'KOSPI_ret', 'BASE_RATE']

for col in analysis_cols:
    if col in df_market.columns:
        valid = df_market[col].notna().sum()
        first_valid = df_market.loc[df_market[col].notna(), 'date'].min()
        last_valid = df_market.loc[df_market[col].notna(), 'date'].max()
        print(f"  {col:15s}: {valid:,}ì¼, {first_valid.strftime('%Y-%m-%d')} ~ {last_valid.strftime('%Y-%m-%d')}")
    else:
        print(f"  {col:15s}: ì»¬ëŸ¼ ì—†ìŒ!")

# ëª¨ë“  ì»¬ëŸ¼ì´ ë™ì‹œì— ìœ íš¨í•œ í–‰ ìˆ˜
print(f"\n  â†’ ëª¨ë“  ì»¬ëŸ¼ ë™ì‹œ ìœ íš¨: {df_market[analysis_cols].dropna().shape[0]}ì¼")

print(f"\n[ë³€ìˆ˜ ì •ì˜] - Gorton (2012) ëŒ€ì‘")
print(f"  RepoSpread   = RPê¸ˆë¦¬ - ê¸°ì¤€ê¸ˆë¦¬ (bp)        â† Repo - OIS")
print(f"  BankStress   = ì½œê¸ˆë¦¬ - ê¸°ì¤€ê¸ˆë¦¬ (bp)        â† LIB-OIS")
print(f"  CreditSpread = íšŒì‚¬ì±„BBB- - êµ­ê³ ì±„3Y (bp)    â† ABX BBB")

# ìƒê´€ê´€ê³„
print(f"\n[í•µì‹¬ ë³€ìˆ˜ ìƒê´€ê´€ê³„ - ì „ì²´ ì‹œì¥]")
corr_cols = ['RepoSpread', 'BankStress', 'CreditSpread', 'VKOSPI']
corr_matrix = df_analysis[corr_cols].corr()
print(corr_matrix.round(3))

print("\nâœ“ ë°ì´í„° ì „ì²˜ë¦¬ ì™„ë£Œ")
```

```python
#%%============================================================================
# [Cell 2.5] ë‚´ìƒì„± ë° ë³€ìˆ˜ ê´€ê³„ ì§„ë‹¨ (íšŒê·€ë¶„ì„ ì „ ì‚¬ì „ ê²€ì¦)
#==============================================================================
import seaborn as sns
import matplotlib.pyplot as plt
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tsa.stattools import grangercausalitytests

# í•œê¸€ í°íŠ¸ ì„¤ì • (Windows ê¸°ì¤€)
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

print("=" * 70)
print("ğŸ”¬ ë‚´ìƒì„±(Endogeneity) ë° ë‹¤ì¤‘ê³µì„ ì„± ì§„ë‹¨")
print("=" * 70)

# ì§„ë‹¨ì— ì‚¬ìš©í•  ë°ì´í„° (ê²°ì¸¡ì¹˜ ì œê±°)
diag_df = df_analysis.copy().dropna()

# ---------------------------------------------------------
# 1. ìƒê´€ê´€ê³„ ë¶„ì„ (Correlation Matrix)
# ---------------------------------------------------------
print("\n[1] ì£¼ìš” ë³€ìˆ˜ ìƒê´€ê³„ìˆ˜ í–‰ë ¬")
target_cols = ['RepoSpread', 'BankStress', 'NonBankStress', 'CreditSpread', 'VKOSPI']
corr_mat = diag_df[target_cols].corr()
print(corr_mat.round(3))

# íˆíŠ¸ë§µ ì‹œê°í™”
plt.figure(figsize=(10, 8))
sns.heatmap(corr_mat, annot=True, fmt='.2f', cmap='RdBu_r', vmin=-1, vmax=1)
plt.title('ì£¼ìš” ë³€ìˆ˜ ê°„ ìƒê´€ê´€ê³„ (Pearson)')
plt.show()

# ---------------------------------------------------------
# 2. ë‹¤ì¤‘ê³µì„ ì„±(VIF) ì§„ë‹¨
# ---------------------------------------------------------
print("\n[2] ë‹¤ì¤‘ê³µì„ ì„±(VIF) ì§„ë‹¨")
print("   â€» VIF > 5: ì£¼ì˜, VIF > 10: ì‹¬ê°í•œ ê³µì„ ì„± (ë³€ìˆ˜ ì œê±° ê³ ë ¤)")

# íšŒê·€ì‹ì— ë“¤ì–´ê°€ëŠ” ë…ë¦½ë³€ìˆ˜ë“¤ (ìƒìˆ˜í•­ í¬í•¨)
X_vif = diag_df[['BankStress', 'CreditSpread', 'VKOSPI', 'KOSPI_ret', 'YieldSlope']]
X_vif = sm.add_constant(X_vif)

vif_data = pd.DataFrame()
vif_data["ë³€ìˆ˜"] = X_vif.columns
vif_data["VIF"] = [variance_inflation_factor(X_vif.values, i) for i in range(X_vif.shape[1])]

print(vif_data.round(2).to_string(index=False))

# ---------------------------------------------------------
# 3. Granger ì¸ê³¼ê´€ê³„ ê²€ì • (ë‚´ìƒì„± í™•ì¸)
# ---------------------------------------------------------
print("\n[3] Granger ì¸ê³¼ê´€ê³„ ê²€ì • (RepoSpreadì™€ì˜ ê´€ê³„)")
print("   â€» ê·€ë¬´ê°€ì„¤: XëŠ” Yë¥¼ ì˜ˆì¸¡í•˜ëŠ” ë° ë„ì›€ì´ ë˜ì§€ ì•ŠëŠ”ë‹¤ (X -> Y ì¸ê³¼ì„± ì—†ìŒ)")
print("   â€» p-value < 0.05 ì´ë©´ ê·€ë¬´ê°€ì„¤ ê¸°ê° (ì¦‰, ì¸ê³¼ì„± ìˆìŒ)")

# Granger ê²€ì •ì€ ì‹œê³„ì—´ì˜ 'ì•ˆì •ì„±'ì´ í•„ìš”í•˜ë¯€ë¡œ 1ì°¨ ì°¨ë¶„ ë°ì´í„° ì‚¬ìš© ê¶Œì¥
diag_diff = diag_df.diff().dropna()

# ì£¼ìš” ë³€ìˆ˜ë“¤ì— ëŒ€í•´ í…ŒìŠ¤íŠ¸ (Max Lag = 3)
test_vars = ['BankStress', 'NonBankStress', 'CreditSpread']

for var in test_vars:
    print(f"\n   Testing: {var} vs RepoSpread")
    
    # 1) X -> Y (ìš°ë¦¬ê°€ ì›í•˜ëŠ” ë°©í–¥)
    # ë°ì´í„° ìˆœì„œ: [Y, X] -> Xê°€ Yì— ì˜í–¥ì„ ì£¼ëŠ”ê°€?
    print(f"   [ë°©í–¥: {var} -> RepoSpread]")
    try:
        gc_res = grangercausalitytests(diag_diff[['RepoSpread', var]], maxlag=[3], verbose=False)
        p_val = gc_res[3][0]['ssr_chi2test'][1] # lag 3ì˜ p-value
        print(f"      Lag 3 p-value: {p_val:.4f} ({'ìœ ì˜í•¨: ì¸ê³¼ì„± O' if p_val < 0.05 else 'ìœ ì˜í•˜ì§€ ì•ŠìŒ'})")
    except Exception as e:
        print(f"      ì˜¤ë¥˜ ë°œìƒ: {e}")

    # 2) Y -> X (ì—­ì¸ê³¼ì„±: RepoSpreadê°€ ë³€í•´ì„œ BankStressê°€ ë³€í–ˆë‚˜? -> ë‚´ìƒì„± ì˜ì‹¬)
    # ë°ì´í„° ìˆœì„œ: [X, Y] -> Yê°€ Xì— ì˜í–¥ì„ ì£¼ëŠ”ê°€?
    print(f"   [ë°©í–¥: RepoSpread -> {var}] (ì—­ì¸ê³¼ì„± ì²´í¬)")
    try:
        gc_res_rev = grangercausalitytests(diag_diff[[var, 'RepoSpread']], maxlag=[3], verbose=False)
        p_val_rev = gc_res_rev[3][0]['ssr_chi2test'][1]
        print(f"      Lag 3 p-value: {p_val_rev:.4f} ({'âš ï¸ ì—­ì¸ê³¼ì„± ì˜ì‹¬!' if p_val_rev < 0.05 else 'ì•ˆì „í•¨'})")
    except Exception as e:
        print(f"      ì˜¤ë¥˜ ë°œìƒ: {e}")

print("-" * 70)
```

```python
#%%============================================================================
# [Cell 3] íšŒê·€ë¶„ì„ ì‹¤í–‰ ë° ê²°ê³¼ ì¶œë ¥
#==============================================================================
"""
Gorton & Metrick (2012) í•œêµ­ ì ìš© - íšŒê·€ë¶„ì„

[ë¶„ì„ êµ¬ì¡°]
Part 1: ì „ì²´ ì‹œì¥ ìˆ˜ì¤€ íšŒê·€ (Repo ê¸°ë°˜)
Part 2: ì „ì²´ ì‹œì¥ 1ì°¨ ì°¨ë¶„ íšŒê·€
Part 3: íŒ¨ë„ ê³ ì •íš¨ê³¼ íšŒê·€ (ë‹´ë³´ìœ í˜•ë³„)
Part 4: íŒ¨ë„ 1ì°¨ ì°¨ë¶„ íšŒê·€
Part 5: ë‹´ë³´ìœ í˜•ë³„ ê°œë³„ íšŒê·€ (NEW)

[ë³€ìˆ˜ êµ¬ì¡°]
- ì¢…ì†ë³€ìˆ˜: RepoSpread = RPê¸ˆë¦¬ - ê¸°ì¤€ê¸ˆë¦¬ (bp)
- BankStress: ì½œê¸ˆë¦¬ - ê¸°ì¤€ê¸ˆë¦¬ (bp) â† LIB-OIS ëŒ€ì‘
- CreditSpread: CP91-í†µì•ˆì±„91 â† ABX ëŒ€ì‘
- VKOSPI: ë³€ë™ì„±ì§€ìˆ˜ â† VIX (í†µì œë³€ìˆ˜)
"""

from linearmodels.panel import PanelOLS

# ì¶œë ¥ í˜•ì‹ ì„¤ì •
pd.set_option('display.float_format', lambda x: '%.4f' % x)
np.set_printoptions(suppress=True, precision=4)

print("=" * 70)
print("ğŸ“Š Gorton & Metrick (2012) ìŠ¤íƒ€ì¼ íšŒê·€ë¶„ì„")
print("=" * 70)
print("""
[ëª¨í˜•]

1) ì „ì²´ ì‹œì¥ (Repo ê¸°ë°˜):
   ìˆ˜ì¤€: RepoSpread_t = Î± + Î²â‚BankStress_t + Î²â‚‚CreditSpread_t + Î³X_t + Îµ_t
   ì°¨ë¶„: Î”RepoSpread_t = Î± + Î£Î²â‚Î”BankStress_{t-k} + Î£Î²â‚‚Î”CreditSpread_{t-k} + Î³Î”X_t + Îµ_t

2) íŒ¨ë„ íšŒê·€ (ë‹´ë³´ìœ í˜•ë³„):
   ìˆ˜ì¤€: RepoSpread_{i,t} = Î±_i + Î²â‚BankStress_t + Î²â‚‚CreditSpread_t + Î³X_t + Îµ_{i,t}
   i = ë‹´ë³´ìœ í˜•, Î±_i = ë‹´ë³´ ê³ ì •íš¨ê³¼

3) ë‹´ë³´ìœ í˜•ë³„ ê°œë³„ íšŒê·€:
   RepoSpread_{i,t} = Î±_i + Î²â‚_i BankStress_t + Î²â‚‚_i CreditSpread_t + Î³_i X_t + Îµ_{i,t}
   ë‹´ë³´ìœ í˜•ë§ˆë‹¤ ë³„ë„ì˜ ê³„ìˆ˜ (Î²â‚_i, Î²â‚‚_i) ì¶”ì •

[ë³€ìˆ˜ ì •ì˜] - Gorton (2012) ëŒ€ì‘
- RepoSpread   = RPê¸ˆë¦¬ - ê¸°ì¤€ê¸ˆë¦¬ (bp)        â† Repo - OIS
- BankStress   = ì½œê¸ˆë¦¬ - ê¸°ì¤€ê¸ˆë¦¬ (bp)        â† LIB-OIS
- CreditSpread = íšŒì‚¬ì±„BBB- - êµ­ê³ ì±„3Y (bp)    â† ABX BBB
""")


def format_pvalue(p):
    """p-value í¬ë§·íŒ…"""
    if p < 0.001:
        return "<0.001"
    else:
        return f"{p:.4f}"


def get_significance(p):
    """ìœ ì˜ìˆ˜ì¤€ ë³„í‘œ"""
    if p < 0.01:
        return '***'
    elif p < 0.05:
        return '**'
    elif p < 0.1:
        return '*'
    return ''


# ========================================
# Part 1: ì „ì²´ ì‹œì¥ ìˆ˜ì¤€ íšŒê·€ (KOFR)
# ========================================

print("\n" + "=" * 70)
print("ğŸ“Œ Part 1: ì „ì²´ ì‹œì¥ ìˆ˜ì¤€ íšŒê·€ (KOFR ê¸°ë°˜)")
print("=" * 70)

# ì¢…ì†ë³€ìˆ˜
y_level = df_analysis['RepoSpread'].astype(float)

# ë…ë¦½ë³€ìˆ˜
X_level_vars = ['BankStress', 'CreditSpread', 'VKOSPI', 'KOSPI_ret', 'YieldSlope', 'r10']
X_level = df_analysis[X_level_vars].astype(float)
X_level = sm.add_constant(X_level)

# OLS íšŒê·€ (HAC í‘œì¤€ì˜¤ì°¨)
model_level = OLS(y_level, X_level).fit(cov_type='HAC', cov_kwds={'maxlags': 5})

print(f"\nê´€ì¸¡ì¹˜: {int(model_level.nobs)}, RÂ²: {model_level.rsquared:.4f}, Adj-RÂ²: {model_level.rsquared_adj:.4f}")

# ê²°ê³¼ ì •ë¦¬
level_results = pd.DataFrame({
    'ë³€ìˆ˜': model_level.params.index,
    'ê³„ìˆ˜': model_level.params.values.round(4),
    'í‘œì¤€ì˜¤ì°¨': model_level.bse.values.round(4),
    't-stat': model_level.tvalues.values.round(3),
    'p-value': [format_pvalue(p) for p in model_level.pvalues],
    'ìœ ì˜': [get_significance(p) for p in model_level.pvalues]
})

print(f"\níšŒê·€ê³„ìˆ˜:")
print(level_results.to_string(index=False))


# ========================================
# Part 2: ì „ì²´ ì‹œì¥ 1ì°¨ ì°¨ë¶„ íšŒê·€
# ========================================

print("\n" + "=" * 70)
print("ğŸ“Œ Part 2: ì „ì²´ ì‹œì¥ 1ì°¨ ì°¨ë¶„ íšŒê·€")
print("=" * 70)

# 1ì°¨ ì°¨ë¶„ ë°ì´í„° ìƒì„±
df_diff = df_analysis.copy()
df_diff['d_spread'] = df_diff['RepoSpread'].diff()
df_diff['d_bank'] = df_diff['BankStress'].diff()
df_diff['d_credit'] = df_diff['CreditSpread'].diff()
df_diff['d_vkospi'] = df_diff['VKOSPI'].diff()
df_diff['d_slope'] = df_diff['YieldSlope'].diff()
df_diff['d_r10'] = df_diff['r10'].diff()

# ì‹œì°¨ ë³€ìˆ˜ ìƒì„±
for lag in range(1, N_LAGS + 1):
    df_diff[f'd_bank_L{lag}'] = df_diff['d_bank'].shift(lag)
    df_diff[f'd_credit_L{lag}'] = df_diff['d_credit'].shift(lag)

df_diff = df_diff.dropna()

# ì¢…ì†ë³€ìˆ˜
y_diff = df_diff['d_spread'].astype(float)

# ë…ë¦½ë³€ìˆ˜
X_diff_vars = ['d_bank', 'd_credit', 'd_vkospi', 'KOSPI_ret', 'd_slope', 'd_r10']
for lag in range(1, N_LAGS + 1):
    X_diff_vars.extend([f'd_bank_L{lag}', f'd_credit_L{lag}'])

X_diff = df_diff[X_diff_vars].astype(float)
X_diff = sm.add_constant(X_diff)

# OLS íšŒê·€
model_diff = OLS(y_diff, X_diff).fit(cov_type='HC1')

print(f"\nê´€ì¸¡ì¹˜: {int(model_diff.nobs)}, RÂ²: {model_diff.rsquared:.4f}, Adj-RÂ²: {model_diff.rsquared_adj:.4f}")

# ê²°ê³¼ ì •ë¦¬
diff_results = pd.DataFrame({
    'ë³€ìˆ˜': model_diff.params.index,
    'ê³„ìˆ˜': model_diff.params.values.round(4),
    'í‘œì¤€ì˜¤ì°¨': model_diff.bse.values.round(4),
    't-stat': model_diff.tvalues.values.round(3),
    'p-value': [format_pvalue(p) for p in model_diff.pvalues],
    'ìœ ì˜': [get_significance(p) for p in model_diff.pvalues]
})

print(f"\níšŒê·€ê³„ìˆ˜:")
print(diff_results.to_string(index=False))

# F-ê²€ì •
print(f"\n[F-ê²€ì •: ë³€ìˆ˜ ê·¸ë£¹ ê³µë™ ìœ ì˜ì„±]")
for var_name, var_prefix in [('BankStress', 'd_bank'), ('CreditSpread', 'd_credit')]:
    vars_list = [c for c in X_diff_vars if c.startswith(var_prefix)]
    if len(vars_list) > 0:
        try:
            r_matrix = np.zeros((len(vars_list), len(X_diff.columns)))
            for i, var in enumerate(vars_list):
                r_matrix[i, list(X_diff.columns).index(var)] = 1
            f_test = model_diff.f_test(r_matrix)
            f_val = float(np.array(f_test.fvalue).flat[0])
            f_p = float(np.array(f_test.pvalue).flat[0])
            sig = get_significance(f_p)
            p_str = "<0.001" if f_p < 0.001 else f"{f_p:.4f}"
            print(f"  {var_name:15s}: F = {f_val:6.2f}, p = {p_str} {sig}")
        except Exception as e:
            print(f"  {var_name}: F-ê²€ì • ì˜¤ë¥˜ ({e})")


# ========================================
# Part 3: íŒ¨ë„ ê³ ì •íš¨ê³¼ íšŒê·€ (ìˆ˜ì¤€)
# ========================================

print("\n" + "=" * 70)
print("ğŸ“Œ Part 3: íŒ¨ë„ ê³ ì •íš¨ê³¼ íšŒê·€ (ë‹´ë³´ìœ í˜•ë³„, ìˆ˜ì¤€)")
print("=" * 70)

# íŒ¨ë„ ë°ì´í„° ì¤€ë¹„
panel_vars = ['RepoSpread', 'BankStress', 'CreditSpread', 'VKOSPI', 'KOSPI_ret', 'YieldSlope', 'r10']
df_panel_reg = df_panel[panel_vars].dropna()

# ì¢…ì†ë³€ìˆ˜
y_panel = df_panel_reg['RepoSpread']

# ë…ë¦½ë³€ìˆ˜
X_panel_vars = ['BankStress', 'CreditSpread', 'VKOSPI', 'KOSPI_ret', 'YieldSlope', 'r10']
X_panel = df_panel_reg[X_panel_vars]

# íŒ¨ë„ ê³ ì •íš¨ê³¼ íšŒê·€
model_panel = PanelOLS(y_panel, X_panel, entity_effects=True).fit(cov_type='clustered', cluster_entity=True)

print(f"\nê´€ì¸¡ì¹˜: {int(model_panel.nobs)}, RÂ² (within): {model_panel.rsquared:.4f}")
print(f"ë‹´ë³´ìœ í˜•: {df_panel_reg.index.get_level_values('collateral').nunique()}ê°œ")

# ê²°ê³¼ ì •ë¦¬
panel_results = pd.DataFrame({
    'ë³€ìˆ˜': model_panel.params.index,
    'ê³„ìˆ˜': model_panel.params.values.round(4),
    'í‘œì¤€ì˜¤ì°¨': model_panel.std_errors.values.round(4),
    't-stat': model_panel.tstats.values.round(3),
    'p-value': [format_pvalue(p) for p in model_panel.pvalues],
    'ìœ ì˜': [get_significance(p) for p in model_panel.pvalues]
})

print(f"\níšŒê·€ê³„ìˆ˜:")
print(panel_results.to_string(index=False))


# ========================================
# Part 4: íŒ¨ë„ 1ì°¨ ì°¨ë¶„ íšŒê·€
# ========================================

print("\n" + "=" * 70)
print("ğŸ“Œ Part 4: íŒ¨ë„ 1ì°¨ ì°¨ë¶„ íšŒê·€ (ë‹´ë³´ìœ í˜•ë³„)")
print("=" * 70)

# íŒ¨ë„ ì°¨ë¶„ ë°ì´í„° ìƒì„±
df_panel_diff = df_panel.copy()

# ë‹´ë³´ìœ í˜•ë³„ë¡œ ì°¨ë¶„ ê³„ì‚°
df_panel_diff['d_spread'] = df_panel_diff.groupby(level='collateral')['RepoSpread'].diff()
df_panel_diff['d_bank'] = df_panel_diff.groupby(level='collateral')['BankStress'].diff()
df_panel_diff['d_credit'] = df_panel_diff.groupby(level='collateral')['CreditSpread'].diff()
df_panel_diff['d_vkospi'] = df_panel_diff.groupby(level='collateral')['VKOSPI'].diff()
df_panel_diff['d_slope'] = df_panel_diff.groupby(level='collateral')['YieldSlope'].diff()
df_panel_diff['d_r10'] = df_panel_diff.groupby(level='collateral')['r10'].diff()

# ì‹œì°¨ ë³€ìˆ˜ ìƒì„±
for lag in range(1, N_LAGS + 1):
    df_panel_diff[f'd_bank_L{lag}'] = df_panel_diff.groupby(level='collateral')['d_bank'].shift(lag)
    df_panel_diff[f'd_credit_L{lag}'] = df_panel_diff.groupby(level='collateral')['d_credit'].shift(lag)

df_panel_diff = df_panel_diff.dropna()

# ì¢…ì†ë³€ìˆ˜
y_panel_diff = df_panel_diff['d_spread']

# ë…ë¦½ë³€ìˆ˜
X_panel_diff_vars = ['d_bank', 'd_credit', 'd_vkospi', 'KOSPI_ret', 'd_slope', 'd_r10']
for lag in range(1, N_LAGS + 1):
    X_panel_diff_vars.extend([f'd_bank_L{lag}', f'd_credit_L{lag}'])

X_panel_diff = df_panel_diff[X_panel_diff_vars]

# íŒ¨ë„ ê³ ì •íš¨ê³¼ íšŒê·€ (ì°¨ë¶„)
model_panel_diff = PanelOLS(y_panel_diff, X_panel_diff, entity_effects=True).fit(cov_type='clustered', cluster_entity=True)

print(f"\nê´€ì¸¡ì¹˜: {int(model_panel_diff.nobs)}, RÂ² (within): {model_panel_diff.rsquared:.4f}")

# ê²°ê³¼ ì •ë¦¬
panel_diff_results = pd.DataFrame({
    'ë³€ìˆ˜': model_panel_diff.params.index,
    'ê³„ìˆ˜': model_panel_diff.params.values.round(4),
    'í‘œì¤€ì˜¤ì°¨': model_panel_diff.std_errors.values.round(4),
    't-stat': model_panel_diff.tstats.values.round(3),
    'p-value': [format_pvalue(p) for p in model_panel_diff.pvalues],
    'ìœ ì˜': [get_significance(p) for p in model_panel_diff.pvalues]
})

print(f"\níšŒê·€ê³„ìˆ˜:")
print(panel_diff_results.to_string(index=False))

# F-ê²€ì • (íŒ¨ë„)
print(f"\n[F-ê²€ì •: ë³€ìˆ˜ ê·¸ë£¹ ê³µë™ ìœ ì˜ì„±]")
for var_name, var_prefix in [('BankStress', 'd_bank'), ('CreditSpread', 'd_credit')]:
    vars_list = [c for c in X_panel_diff_vars if c.startswith(var_prefix)]
    if len(vars_list) > 0:
        try:
            # Wald ê²€ì •
            restriction = ' = '.join([f'{v}' for v in vars_list]) + ' = 0'
            wald_test = model_panel_diff.wald_test(formula=restriction)
            f_val = float(wald_test.stat)
            f_p = float(wald_test.pval)
            sig = get_significance(f_p)
            p_str = "<0.001" if f_p < 0.001 else f"{f_p:.4f}"
            print(f"  {var_name:15s}: Wald = {f_val:6.2f}, p = {p_str} {sig}")
        except Exception as e:
            print(f"  {var_name}: Wald ê²€ì • ì˜¤ë¥˜ ({e})")


# ========================================
# Part 5: ë‹´ë³´ìœ í˜•ë³„ ê°œë³„ íšŒê·€
# ========================================

print("\n" + "=" * 70)
print("ğŸ“Œ Part 5: ë‹´ë³´ìœ í˜•ë³„ ê°œë³„ íšŒê·€ (ìˆ˜ì¤€)")
print("=" * 70)
print("\nê° ë‹´ë³´ìœ í˜•ì— ëŒ€í•´ ë³„ë„ íšŒê·€ â†’ ë‹´ë³´ë³„ ë¯¼ê°ë„(Î²) ì¶”ì •\n")

# ë‹´ë³´ìœ í˜• ë¦¬ìŠ¤íŠ¸
collateral_types = df_panel.index.get_level_values('collateral').unique().tolist()

# ê²°ê³¼ ì €ì¥ìš©
collateral_level_results = []

for coll in sorted(collateral_types):
    # í•´ë‹¹ ë‹´ë³´ ë°ì´í„° ì¶”ì¶œ
    df_coll = df_panel.xs(coll, level='collateral').copy()
    df_coll = df_coll[panel_vars].dropna()
    
    if len(df_coll) < 50:
        continue
    
    # íšŒê·€ë¶„ì„
    y = df_coll['RepoSpread'].astype(float)
    X = df_coll[X_panel_vars].astype(float)
    X = sm.add_constant(X)
    
    try:
        model = OLS(y, X).fit(cov_type='HAC', cov_kwds={'maxlags': 5})
        
        collateral_level_results.append({
            'ë‹´ë³´ìœ í˜•': coll,
            'N': int(model.nobs),
            'RÂ²': round(model.rsquared, 4),
            'BankStress': model.params.get('BankStress', np.nan),
            'BankStress_p': model.pvalues.get('BankStress', np.nan),
            'CreditSpread': model.params.get('CreditSpread', np.nan),
            'CreditSpread_p': model.pvalues.get('CreditSpread', np.nan),
            'VKOSPI': model.params.get('VKOSPI', np.nan),
            'VKOSPI_p': model.pvalues.get('VKOSPI', np.nan)
        })
    except Exception as e:
        print(f"  {coll}: íšŒê·€ ì˜¤ë¥˜ ({e})")

# ê²°ê³¼ ì¶œë ¥
print("[ë‹´ë³´ìœ í˜•ë³„ ìˆ˜ì¤€ íšŒê·€ ê²°ê³¼]")
print("-" * 90)

coll_level_df = pd.DataFrame(collateral_level_results)

# í¬ë§·íŒ…
def format_coef(row, var):
    coef = row[var]
    pval = row[f'{var}_p']
    sig = get_significance(pval)
    return f"{coef:7.4f}{sig}"

coll_level_summary = coll_level_df[['ë‹´ë³´ìœ í˜•', 'N', 'RÂ²']].copy()
coll_level_summary['BankStress'] = coll_level_df.apply(lambda r: format_coef(r, 'BankStress'), axis=1)
coll_level_summary['CreditSpread'] = coll_level_df.apply(lambda r: format_coef(r, 'CreditSpread'), axis=1)
coll_level_summary['VKOSPI'] = coll_level_df.apply(lambda r: format_coef(r, 'VKOSPI'), axis=1)

print(coll_level_summary.to_string(index=False))


# ========================================
# Part 6: ë‹´ë³´ìœ í˜•ë³„ ê°œë³„ íšŒê·€ (1ì°¨ ì°¨ë¶„)
# ========================================

print("\n" + "=" * 70)
print("ğŸ“Œ Part 6: ë‹´ë³´ìœ í˜•ë³„ ê°œë³„ íšŒê·€ (1ì°¨ ì°¨ë¶„)")
print("=" * 70)

# ê²°ê³¼ ì €ì¥ìš©
collateral_diff_results = []

for coll in sorted(collateral_types):
    # í•´ë‹¹ ë‹´ë³´ ë°ì´í„° ì¶”ì¶œ
    df_coll = df_panel.xs(coll, level='collateral').copy()
    df_coll = df_coll[panel_vars].dropna()
    
    if len(df_coll) < 50:
        continue
    
    # 1ì°¨ ì°¨ë¶„
    df_coll['d_spread'] = df_coll['RepoSpread'].diff()
    df_coll['d_bank'] = df_coll['BankStress'].diff()
    df_coll['d_credit'] = df_coll['CreditSpread'].diff()
    df_coll['d_vkospi'] = df_coll['VKOSPI'].diff()
    df_coll['d_slope'] = df_coll['YieldSlope'].diff()
    df_coll['d_r10'] = df_coll['r10'].diff()
    
    # ì‹œì°¨ ë³€ìˆ˜
    for lag in range(1, N_LAGS + 1):
        df_coll[f'd_bank_L{lag}'] = df_coll['d_bank'].shift(lag)
        df_coll[f'd_credit_L{lag}'] = df_coll['d_credit'].shift(lag)
    
    df_coll = df_coll.dropna()
    
    if len(df_coll) < 50:
        continue
    
    # íšŒê·€ë¶„ì„
    y = df_coll['d_spread'].astype(float)
    X_vars_diff = ['d_bank', 'd_credit', 'd_vkospi', 'KOSPI_ret', 'd_slope', 'd_r10']
    for lag in range(1, N_LAGS + 1):
        X_vars_diff.extend([f'd_bank_L{lag}', f'd_credit_L{lag}'])
    
    X = df_coll[X_vars_diff].astype(float)
    X = sm.add_constant(X)
    
    try:
        model = OLS(y, X).fit(cov_type='HC1')
        
        # BankStress í•©ê³„ (ë™ì‹œ + ì‹œì°¨)
        bank_vars = ['d_bank'] + [f'd_bank_L{i}' for i in range(1, N_LAGS + 1)]
        bank_sum = sum([model.params.get(v, 0) for v in bank_vars])
        
        # CreditSpread í•©ê³„
        credit_vars = ['d_credit'] + [f'd_credit_L{i}' for i in range(1, N_LAGS + 1)]
        credit_sum = sum([model.params.get(v, 0) for v in credit_vars])
        
        collateral_diff_results.append({
            'ë‹´ë³´ìœ í˜•': coll,
            'N': int(model.nobs),
            'RÂ²': round(model.rsquared, 4),
            'd_bank': model.params.get('d_bank', np.nan),
            'd_bank_p': model.pvalues.get('d_bank', np.nan),
            'd_credit': model.params.get('d_credit', np.nan),
            'd_credit_p': model.pvalues.get('d_credit', np.nan),
            'd_vkospi': model.params.get('d_vkospi', np.nan),
            'd_vkospi_p': model.pvalues.get('d_vkospi', np.nan),
            'BankStress_ì´í•©': bank_sum,
            'CreditSpread_ì´í•©': credit_sum
        })
    except Exception as e:
        print(f"  {coll}: íšŒê·€ ì˜¤ë¥˜ ({e})")

# ê²°ê³¼ ì¶œë ¥
print("\n[ë‹´ë³´ìœ í˜•ë³„ 1ì°¨ ì°¨ë¶„ íšŒê·€ ê²°ê³¼]")
print("-" * 90)

coll_diff_df = pd.DataFrame(collateral_diff_results)

coll_diff_summary = coll_diff_df[['ë‹´ë³´ìœ í˜•', 'N', 'RÂ²']].copy()
coll_diff_summary['d_bank(t)'] = coll_diff_df.apply(lambda r: format_coef(r, 'd_bank'), axis=1)
coll_diff_summary['d_credit(t)'] = coll_diff_df.apply(lambda r: format_coef(r, 'd_credit'), axis=1)
coll_diff_summary['d_vkospi'] = coll_diff_df.apply(lambda r: format_coef(r, 'd_vkospi'), axis=1)
coll_diff_summary['Bank_ì´í•©'] = coll_diff_df['BankStress_ì´í•©'].round(4)
coll_diff_summary['Credit_ì´í•©'] = coll_diff_df['CreditSpread_ì´í•©'].round(4)

print(coll_diff_summary.to_string(index=False))


# ========================================
# Part 7: ê²°ê³¼ ìš”ì•½
# ========================================

print(f"\n{'='*70}")
print("ğŸ“Š íšŒê·€ë¶„ì„ ê²°ê³¼ ìš”ì•½")
print("="*70)

print("\n[1] ì „ì²´ ì‹œì¥ - ìˆ˜ì¤€ íšŒê·€ (KOFR ê¸°ë°˜)")
print("-" * 70)
print(f"RÂ² = {model_level.rsquared:.4f}, Adj-RÂ² = {model_level.rsquared_adj:.4f}")
key_vars = ['const', 'BankStress', 'CreditSpread', 'VKOSPI']
print(level_results[level_results['ë³€ìˆ˜'].isin(key_vars)].to_string(index=False))

print("\n\n[2] ì „ì²´ ì‹œì¥ - 1ì°¨ ì°¨ë¶„ íšŒê·€")
print("-" * 70)
print(f"RÂ² = {model_diff.rsquared:.4f}, Adj-RÂ² = {model_diff.rsquared_adj:.4f}")
key_vars = ['const', 'd_bank', 'd_credit', 'd_vkospi']
print(diff_results[diff_results['ë³€ìˆ˜'].isin(key_vars)].to_string(index=False))

print("\n\n[3] íŒ¨ë„ - ìˆ˜ì¤€ íšŒê·€ (ë‹´ë³´ ê³ ì •íš¨ê³¼, ê³µí†µ ê³„ìˆ˜)")
print("-" * 70)
print(f"RÂ² (within) = {model_panel.rsquared:.4f}")
key_vars = ['BankStress', 'CreditSpread', 'VKOSPI']
print(panel_results[panel_results['ë³€ìˆ˜'].isin(key_vars)].to_string(index=False))

print("\n\n[4] íŒ¨ë„ - 1ì°¨ ì°¨ë¶„ íšŒê·€ (ë‹´ë³´ ê³ ì •íš¨ê³¼, ê³µí†µ ê³„ìˆ˜)")
print("-" * 70)
print(f"RÂ² (within) = {model_panel_diff.rsquared:.4f}")
key_vars = ['d_bank', 'd_credit', 'd_vkospi']
print(panel_diff_results[panel_diff_results['ë³€ìˆ˜'].isin(key_vars)].to_string(index=False))

print("\n\n[5] ë‹´ë³´ìœ í˜•ë³„ ê°œë³„ íšŒê·€ - ìˆ˜ì¤€")
print("-" * 70)
print(coll_level_summary.to_string(index=False))

print("\n\n[6] ë‹´ë³´ìœ í˜•ë³„ ê°œë³„ íšŒê·€ - 1ì°¨ ì°¨ë¶„")
print("-" * 70)
print(coll_diff_summary.to_string(index=False))


# ========================================
# Part 8: í•´ì„ ê°€ì´ë“œ
# ========================================

print(f"\n{'='*70}")
print("ğŸ“ ê²°ê³¼ í•´ì„ ê°€ì´ë“œ")
print("="*70)
print("""
[Gorton & Metrick (2012) ëŒ€ì‘]

| Gorton ì›ë…¼ë¬¸           | ë³¸ ì—°êµ¬ (í•œêµ­)               | 
|-------------------------|------------------------------|
| Repo Rate - OIS         | RPê¸ˆë¦¬ - ê¸°ì¤€ê¸ˆë¦¬            |
| LIB-OIS                 | BankStress (ì½œ-ê¸°ì¤€ê¸ˆë¦¬)     |
| ABX BBB                 | íšŒì‚¬ì±„BBB- - êµ­ê³ ì±„3Y        |
| VIX                     | VKOSPI                       |


[ë¶„ì„ êµ¬ì¡° ë¹„êµ]

| ë¶„ì„            | ì¢…ì†ë³€ìˆ˜      | ê³„ìˆ˜      | ì¥ì                     |
|-----------------|---------------|-----------|-------------------------|
| ì „ì²´ ì‹œì¥       | KOFR-ê¸°ì¤€ê¸ˆë¦¬ | ë‹¨ì¼      | ê³µì‹ ë°ì´í„°, ê¸°ì¤€ì      |
| íŒ¨ë„ ê³ ì •íš¨ê³¼   | ë‹´ë³´ë³„ ìŠ¤í”„ë ˆë“œ| ê³µí†µ      | êµ¬ì„±íš¨ê³¼ í†µì œ, ê´€ì¸¡ì¹˜â†‘  |
| ë‹´ë³´ë³„ ê°œë³„íšŒê·€ | ë‹´ë³´ë³„ ìŠ¤í”„ë ˆë“œ| ë‹´ë³´ë³„    | ë¯¼ê°ë„ ì°¨ì´ í¬ì°©        |


[ë‹´ë³´ìœ í˜•ë³„ ë¯¼ê°ë„ í•´ì„]

BankStress ê³„ìˆ˜ê°€ ë‹´ë³´ìœ í˜•ë³„ë¡œ ë‹¤ë¥´ë‹¤ë©´:
- êµ­ê³ ì±„: ë‚®ì€ Î² â†’ ì•ˆì „ìì‚°, ìœ ë™ì„± ìœ„í—˜ì— ëœ ë¯¼ê°
- íšŒì‚¬ì±„: ë†’ì€ Î² â†’ ìœ„í—˜ìì‚°, ìœ ë™ì„± ìœ„í—˜ì— ë” ë¯¼ê°

â†’ "Flight to Quality" í˜„ìƒ í¬ì°© ê°€ëŠ¥


[ê³„ìˆ˜ í•´ì„]

1. BankStress (ì½œê¸ˆë¦¬ - ê¸°ì¤€ê¸ˆë¦¬)
   - ì–‘(+): ì€í–‰ê°„ ìê¸ˆì‹œì¥ ê¸´ì¥ â†’ RP ìŠ¤í”„ë ˆë“œ í™•ëŒ€
   - ë‹´ë³´ë³„ ì°¨ì´: ìœ„í—˜ë‹´ë³´ì¼ìˆ˜ë¡ ë” ë¯¼ê°í•  ê²ƒìœ¼ë¡œ ì˜ˆìƒ

2. CreditSpread (íšŒì‚¬ì±„BBB- - êµ­ê³ ì±„3Y)
   - ì–‘(+): ì‹ ìš©ìœ„í—˜ ì¦ê°€ â†’ RP ìŠ¤í”„ë ˆë“œ í™•ëŒ€
   - ë‹´ë³´ ê°€ì¹˜ ë¶ˆí™•ì‹¤ì„± ì±„ë„
""")


# CSV ì €ì¥
level_results.to_csv(f"{OUTPUT_DIR}/reg_level_kofr.csv", encoding='utf-8-sig', index=False)
diff_results.to_csv(f"{OUTPUT_DIR}/reg_diff_kofr.csv", encoding='utf-8-sig', index=False)
panel_results.to_csv(f"{OUTPUT_DIR}/reg_panel_level.csv", encoding='utf-8-sig', index=False)
panel_diff_results.to_csv(f"{OUTPUT_DIR}/reg_panel_diff.csv", encoding='utf-8-sig', index=False)
coll_level_df.to_csv(f"{OUTPUT_DIR}/reg_collateral_level.csv", encoding='utf-8-sig', index=False)
coll_diff_df.to_csv(f"{OUTPUT_DIR}/reg_collateral_diff.csv", encoding='utf-8-sig', index=False)

print(f"\nâœ“ ê²°ê³¼ ì €ì¥ ì™„ë£Œ: {OUTPUT_DIR}/")
```
